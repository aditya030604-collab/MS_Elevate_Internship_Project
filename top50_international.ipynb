```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("top50 (1).csv", encoding='latin1')

df.head()
```

```python
df = df.rename(columns={"Valence.": "Valence"})
```

```python
def get_sentiment(valence):

    if valence >= 60:
        return "Positive"

    elif valence <= 40:
        return "Negative"

    else:
        return "Neutral"

df["Sentiment"] = df["Valence"].apply(get_sentiment)

df.head()
```

```python
sentiment_counts = df["Sentiment"].value_counts()

print(sentiment_counts)
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Sentiment", data=df)

plt.title("Music Sentiment Distribution")
plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.countplot(x="Genre", hue="Sentiment", data=df)

plt.xticks(rotation=90)

plt.title("Genre vs Sentiment")
plt.show()
```

```python
top_songs = df.sort_values(by="Popularity", ascending=False)

print(top_songs[["Track.Name","Artist.Name","Popularity"]].head(10))
```
